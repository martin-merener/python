```python
import numpy as np
```

### Generate data in 2 dimensions

```python
d = 2
K = 8 # number of classes
n1,n2 = 100,300 # range for number of points per cluster

C = np.random.rand(K,d) # centers of classes

X = np.empty((0,d)) # training data
y = []
nks = []
for k in range(K):
    n_k = np.random.randint(n1,n2) # number of points in cluster k
    nks.append(n_k)
    print(n_k)
    for j in range(n_k):
        x = np.random.normal(C[k,:],0.07).reshape(1,d) # a gaussian perturbation of C[k,:] with standard deviation 0.01
        X = np.concatenate((X,x), axis=0)
        y.append(k)
```

```python
print(nks)
```

### Visualize the data

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
x1 = X[:,0]
x2 = X[:,1]
color = [v/K for v in y]
rgb = plt.get_cmap('jet')(color)
ax.scatter(x1, x2, s=2, color=rgb)
ax.set_xlim([min(x1),max(x1)])
ax.set_ylim([min(x2),max(x2)])
plt.show()
```

### Learning Vector Quantization ---- LVQ

```python
R = 3 # number of prototypes from each class
prototypes = {}
for k in range(K):
    ixs_k = np.random.randint(0,nks[k],R)
    Z = X[np.array(y)==k,:]
    prototypes[k] = Z[ixs_k,:]
```

```python
# PLOTTING THE PROTOTYPES
P = np.empty((0,d))
P_colors = []
for k in range(K):
    P = np.concatenate((P,prototypes[k]), axis=0)
    P_colors.extend([k]*R)
color = [v/K for v in P_colors]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
p1 = P[:,0]
p2 = P[:,1]
rgb = plt.get_cmap('jet')(color)
ax.scatter(p1, p2, s=10, color=rgb)
ax.set_xlim([min(x1),max(x1)])
ax.set_ylim([min(x2),max(x2)])
plt.show()
```

```python
def find_closest(prototypes, x):
    d_closest = np.Inf
    for k in range(K):
        for j in range(R):
            a = np.linalg.norm(x-prototypes[k][j,:])
            if a<d_closest:
                d_closest=a
                opt = (j,k)
    return opt
```

```python
# CHECKING IT FINDS THE CLOSEST
x = X[np.random.randint(0,X.shape[0]),]
pair = find_closest(prototypes, x)
print(x,prototypes[pair[1]][pair[0],:])
```

```python
I = 10000
eps = 1
for i in range(I):
    l = np.random.randint(0,X.shape[0])
    x = X[l,]
    pair = find_closest(prototypes, x)
    prot = prototypes[pair[1]][pair[0],:]
    if y[l]==pair[1]:
        prototypes[pair[1]][pair[0],:] = prot + eps*(x-prot)
    else:
        prototypes[pair[1]][pair[0],:] = prot - eps*(x-prot)
    eps = eps*0.99
```

```python
P = np.empty((0,d))
P_colors = []
for k in range(K):
    P = np.concatenate((P,prototypes[k]), axis=0)
    P_colors.extend([k]*R)
color = [v/K for v in P_colors]

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
p1 = P[:,0]
p2 = P[:,1]
rgb = plt.get_cmap('jet')(color)
ax.scatter(p1, p2, s=10, color=rgb)
ax.set_xlim([min(x1),max(x1)])
ax.set_ylim([min(x2),max(x2)])
plt.show()
```
